```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

#read the Es_FBG_FHD.csv file
df1 = pd.read_csv('Es_FBG_SNN.csv', names = ['${E_x}^{FBG-SNN}$', '${E_y}^{FBG-SNN}$', '${E_z}^{FBG-SNN}$'])
df2 = pd.read_csv('Es_SNN_FHD.csv', names = ['${E_x}^{SNN-FHD}$', '${E_y}^{SNN-FHD}$', '${E_z}^{SNN-FHD}$'])
df3 = pd.read_csv('Es_FBG_FHD.csv', names = ['${E_x}^{FBG-FHD}$', '${E_y}^{FBG-FHD}$', '${E_z}^{FBG-FHD}$'])

```

```python
#plot the data
fig, ax = plt.subplots(figsize=(11.25,7.5))
#sns.set(font_scale=1.5, rc={'text.usetex' : True})
plt.style.use('seaborn-v0_8')
#create an horizontl violin plot of df 
ax1=sns.violinplot(data=df1, ax=ax, orient='h', palette='Greens',inner='box')
for patch in ax1.collections:
    patch.set_alpha(0.8)
ax2=sns.violinplot(data=df2, ax=ax, orient='h', palette='Reds',inner='box')
for patch in ax2.collections:
    patch.set_alpha(0.8)
ax3=sns.violinplot(data=df3, ax=ax, orient='h', palette='Blues',inner='box')
for patch in ax3.collections:
    patch.set_alpha(0.8)

#xlim from -70 to 70
plt.xlim(-70,70)
#xticks from -70 to 70 with step 10
plt.xticks(np.arange(-70, 71, 10))
#grid on the background
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='black')

#export the plot to image with transparent background
plt.savefig('violinplot1.svg', dpi=300, transparent=True)
```
